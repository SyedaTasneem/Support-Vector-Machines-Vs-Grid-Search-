# Support Vector Machines

```python
### Loading the required libraries...

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
### loading the dataset...
df= pd.read_csv('voice-classification.csv')
df.head()
```

```python
#### getting more info on the variables....
df.info()
```

```python
#### Getting a statistical summary of the dataset...
df.describe()
```

```python
### listing the variables(columns)
df.columns
```

```python
### checking for null values....
df.isnull().sum()
Null = df.isnull().sum()
print(Null[Null>0])
```

```python
### checking the dimensions of the dataset...
print('The dimensions of the dataset:', df.shape)
print ('Total number of Labels {}'.format(df.shape[0]))
print( "Total number of males {}".format(df[df.label=='male'].shape[0]))
print ('Total number of females{}'.format(df[df.label == 'female'].shape[0]))
```

```python
### Encoding the target Variable 'label' to binary (0,1)
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
```

```python
df['label'] = encode.fit_transform(df['label'])
df['label'].unique()
```

```python
## Extracting the target variable:
target = df['label']
target.unique()
```

```python
###  Dropping the target variable from the dataset.
df.drop('label',axis =1,inplace = True)
df.head()
```

```python
### Creating x(predictor ) nad y(target) variables...
x = df.values
y = target.values
```

```python
### Importing standard scaler from sklearn n t
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```python
### Importing train_test_split  and LabelEncoder from sklearn ..... 
from sklearn.model_selection import train_test_split
```

```python
#### splitting the dataset ...
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.3, random_state = 100)
xtrain.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import  classification_report
from sklearn.metrics import accuracy_score
```

```python
### fitting the data onto the classifier...
svm = SVC()
svm.fit(xtrain,ytrain)
```

```python
#### prediction
ypred = svm.predict(xtest)
```

```python
### model accuracy
Accuracy = accuracy_score(ypred,ytest)
print(Accuracy)
```

```python
#### classification report:
print(classification_report(ypred,ytest))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001]}
grid =   GridSearchCV(SVC(), param_grid,refit = True, verbose = 2) 
grid.fit(xtrain,ytrain)
```

```python
###  prediction: 
y_pred = grid.predict(xtest)
```

```python
### model accuracy
Accuracy = accuracy_score(y_pred,ytest)
print(Accuracy)
```

```python
#### classification report:
print(classification_report(y_pred,ytest))
```

```python
# Hence both SVM and Grid search seem have almost same prediction accuracy , lets test these classifications for another
# dataset(college classification)
```
